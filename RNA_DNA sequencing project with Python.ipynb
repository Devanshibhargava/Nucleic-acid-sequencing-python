## 2. Reverse, Complement and Reverse-Complement Sequences 

**Following are the nucleic acids in DNA & RNA: -**

    Adenine (A), Thymine (T), Cytosine (C), Guanine (G), Uracil (U)

**Given below are the complementary base pairing rules for DNA and RNA respectively with the only difference of Adenine (A) is complement with Thymine (T) in DNA, whereas Adenine (A) is complement with Uracil (U) in RNA: -** 

    DNA complementary base pairing rule
    A -> T
    C -> G
    G -> C
    T -> A

    RNA complementary base pairing rule
    A -> U
    C -> G
    G -> C
    T -> A

### 2.1. DNA reverse, complement and reverse-complement sequence

```python
temp_DNA = "ATGCCGCTAAACTGACATXXTCAGATC"    # Template DNA Sequence
reve_DNA = ''                               # Reverse DNA Sequence
comp_DNA = ''                               # Complementary DNA Sequence
reve_comp_DNA = ''                          # Reverse-Complement DNA Sequence
```

```python
reve_DNA = temp_DNA[::-1]
reve_DNA
```

```python
for i in temp_DNA:
    if i == "A":
        comp_DNA = comp_DNA + 'T'
    elif i == "T":
        comp_DNA = comp_DNA + 'A'
    elif i == "G":
        comp_DNA = comp_DNA + 'C'
    elif i == "C":
        comp_DNA = comp_DNA + 'G'
    else:
        comp_DNA = comp_DNA + 'X'
print(temp_DNA)
print(comp_DNA)        
```

```python
reve_comp_DNA_1 = comp_DNA[::-1]
reve_comp_DNA_1
```

```python
for i in reve_DNA:
    if i == "A":
        reve_comp_DNA = reve_comp_DNA + 'T'
    elif i == "T":
        reve_comp_DNA = reve_comp_DNA + 'A'
    elif i == "G":
        reve_comp_DNA = reve_comp_DNA + 'C'
    elif i == "C":
        reve_comp_DNA = reve_comp_DNA + 'G'
    else:
        reve_comp_DNA = reve_comp_DNA + 'X'
print("Template DNA strand's sequence is ",temp_DNA)
print("Complementary DNA strand's sequence is ",comp_DNA)
print("Reverse-complementary DNA strand's sequence is ",reve_comp_DNA)
```

### 2.2. RNA reverse, complement and reverse-complement sequence

```python
temp_DNA = "ATGCCGCTAAACTGACATXXTCAGATC"    # Template DNA Sequence
temp_RNA = "AUGCCGCUAAACUGACAUXXUCAGAUC"    # use this sequence for the Revers    
reve_Sequence = ''                          # Reverse RNA Sequence
comp_RNA = ''                               # Complementary RNA Sequence
reve_comp_RNA = ''                          # Reverse-Complement RNA Sequence

#===== a- Reverse RNA Sequence ===============================================
# You need to write code here to reverse the RNA sequenc
reve_Sequence = temp_RNA[::-1]


print("Template RNA strand's sequence is ",temp_RNA)
print("Reverse RNA strand's sequence is ",reve_Sequence)
#=============================================================================

#===== b- Complementary RNA Sequence =========================================
comp_RNA = ""                               # Make sure the comp_RNA is empty
# We need 'for' loop to read the template DNA sequence and 'if'
# statments to add the complement base to the 'comp_RNA'
for i in temp_DNA:
    if i == "A":
        comp_RNA = comp_RNA + 'U'
    elif i == "T":
        comp_RNA = comp_RNA + 'A'
    elif i == "G":
        comp_RNA = comp_RNA + 'C'
    elif i == "C":
        comp_RNA = comp_RNA + 'G'
    else:
        comp_RNA = comp_RNA + 'X'
print("Template DNA strand's sequence is ",temp_DNA)
print("Complementary RNA strand's sequence is ",comp_RNA)
# ============================================================================

# c- Reverse-Complement RNA Sequence =========================================
reve_comp_RNA = ""                          # Make sure reve_comp_RNA is empty
# Like the previous code but in reverse order.
for i in temp_DNA[::-1]:
    if i == "A":
        reve_comp_RNA = reve_comp_RNA + 'U'
    elif i == "T":
        reve_comp_RNA = reve_comp_RNA + 'A'
    elif i == "G":
        reve_comp_RNA = reve_comp_RNA + 'C'
    elif i == "C":
        reve_comp_RNA = reve_comp_RNA + 'G'
    else:
        reve_comp_RNA = reve_comp_RNA + 'X'
print("Template DNA strand's sequence is ",temp_DNA)
print("Reverse_complementary RNA strand's sequence is ",reve_comp_RNA)
# ============================================================================
```

## 3. Project: Complete Genome of Corona Virus (SARS-CoV-2)

From the NCBI website we can get the latest nucleotide records of [SARS-CoV-2](https://www.ncbi.nlm.nih.gov/sars-cov-2/). For this project we are going to work with the Wuhan isolate with NCBI refernce number [NC_045512.2](https://www.ncbi.nlm.nih.gov/nuccore/NC_045512).

```python
import tkinter
import tkinter.filedialog        
root = tkinter.Tk()
root.attributes('-alpha',0.3)
root.withdraw


temp_DNA = ''           # One line of Template DNA Sequence
comp_RNA = ''           # One line of Complementary RNA Sequence
covid_cDNA = ''         # Complete coronavirus cDNA
covid_RNA = ''          # Complete coronavirus RNA

#===== Retrive RNA from the cDNA Sequence =====================================
output_file = open('.\covid_sequence.txt', 'w')

file_path = tkinter.filedialog.askopenfilename(parent=root)
root.destroy()

with open(file_path,'r') as input_data:
    header = input_data.readline().strip()    
    output_file.write(header + ' RNA sequence')
    
    for line in input_data:
        temp_DNA = line.strip()
        # Uncomment the following line to create a variable for the cDNA sequence
        covid_cDNA = covid_cDNA + temp_DNA 
        comp_RNA = ""
        for base in temp_DNA:
            if base == "A":
                  comp_RNA = comp_RNA + 'U'
            elif base == "C":
                comp_RNA = comp_RNA + 'G'
            elif base == "G":
                comp_RNA = comp_RNA + 'C'
            elif base == "T":
                comp_RNA = comp_RNA + 'A'
            else:
                comp_RNA = comp_RNA + '?'
                 
        covid_RNA = covid_RNA + comp_RNA
        output_file.write("\n" + comp_RNA)
        output_file.flush()
        #  Uncomment the print lines to see the data line by line
        print('Template DNA ', temp_DNA)        
        print('Complementary RNA ',comp_RNA)
        print("===================================================")
        
input_data.close()
output_file.close()
#==============================================================================
```
